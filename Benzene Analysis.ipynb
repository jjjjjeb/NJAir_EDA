```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
benzenepath = "Database/Benzene/Benzene.csv"
benzene = pd.read_csv(benzenepath)
benzene.head()
# benzene.columns
```

```python
benzene.info()
```

```python
benzene.describe()
```

```python
columns=['PROGRAM', 'YEAR', 'QUARTER',
       'SAMPLE_DATE', 'SAMPLE_START_TIME', 'DURATION_DESC',
       'SAMPLE_VALUE_REPORTED', 'AQS_UNIT_CODE', 'UNIT_DESC',
       'SAMPLING_FREQUENCY_CODE', 'SAMPLE_VALUE_STD_FINAL_UG_M3',
       'SAMPLE_VALUE_STD_FINAL_TYPE', 'MDL_STD_UG_M3',
       'MDL_TYPE', 'CENSUS_TRACT_POPULATION_2010', 'MONITOR_LATITUDE', 'MONITOR_LONGITUDE']
benzene_clean = benzene.filter(items=columns)
```

```python
benzene_clean.head()
```

```python
benzene_clean['PROGRAM'].value_counts()
```

```python
benzene_clean['DURATION_DESC'].value_counts()
```

```python
benzene_clean['MONITOR_LATITUDE'].value_counts()
```

```python
benzene_clean['CENSUS_TRACT_POPULATION_2010'].value_counts()
```

```python
benzene_hourly = benzene_clean[benzene_clean['DURATION_DESC'] == '1 HOUR']
benzene_hourly.head()
```

```python
benzene_hourly['PROGRAM'].value_counts()
```

```python
benzene_daily = benzene_clean[benzene_clean['DURATION_DESC'] == '24 HOURS']
benzene_daily.head()
```

```python
benzene_daily['PROGRAM'].value_counts()
```

```python
benzene_three = benzene_clean[benzene_clean['DURATION_DESC'] == '3 HOURS']
benzene_three.head()
```

```python
benzene_three['PROGRAM'].value_counts()
```

```python
year_group = benzene_clean.groupby(by='YEAR')
```

```python
year_group['SAMPLE_VALUE_REPORTED'].mean()
```

```python
year_list = year_group['YEAR'].unique()
# year_list
```

```python
plt.figure(figsize=(15,10))
plt.bar(year_list,year_group['SAMPLE_VALUE_REPORTED'].mean(), alpha=0.7)
plt.xlim(1989,2017)
plt.title('Sample Reported Mean by Year', size=20)
plt.xlabel('Year', size=16)
plt.ylabel('Sample Value Mean', size=16)
```

```python
latitude_group = benzene_clean.groupby(by="MONITOR_LATITUDE")
```

```python
latitude_group['SAMPLE_VALUE_REPORTED'].mean()
```

```python
latitude_list = latitude_group['MONITOR_LATITUDE'].unique()
latitude_list
```

```python
plt.figure(figsize=(20,10))
plt.bar(latitude_list, latitude_group['SAMPLE_VALUE_REPORTED'].mean(), edgecolor='k', width =0.1, align='center', alpha=0.5)
plt.title('Sample Reported Mean by Latitude', size=20)
plt.xlabel('Latitude', size=16)
plt.ylabel('Sample value reported mean', size=16)
plt.xlim(39.75,41)
```

```python
# benzene_clean[['Day','Month','Year Abbrev']] = benzene_clean['SAMPLE_DATE'].str.split('-')
benzene_clean['Day'], benzene_clean['Month'], benzene_clean['Year Abbrev'] = benzene_clean['SAMPLE_DATE'].str.split('-', 2).str
benzene_clean.head()
```

```python
month_group = benzene_clean.groupby('Month')
month_group['SAMPLE_VALUE_REPORTED'].mean()
```

```python
month_list = ['Apr','Aug','Dec','Feb','Jan','Jul','Jun','Mar','May','Nov','Oct','Sep']
month_list
```

```python
plt.figure(figsize=(15,10))
plt.bar(month_list,month_group['SAMPLE_VALUE_REPORTED'].mean(), alpha=0.7)
# plt.xlim(1989,2017)
plt.title('Sample Reported Mean by Month', size=20)
plt.xlabel('Month', size=16)
plt.ylabel('Sample Value Mean', size=16)
```
