```python
grouped_data= chromium.groupby(["YEAR","PROGRAM"])

# Mean Tumor Volume from "Tumor Volume (mm3)" in group data
mean=grouped_data["SAMPLE_VALUE_REPORTED"].mean()

# Convert to DataFrame
toxin_df = pd.DataFrame(mean).reset_index()

# Preview DataFrame
toxin_df
```

```python
se_toxin_df = grouped_data["SAMPLE_VALUE_REPORTED"].sem()

# Convert to DataFrame
se_toxin_df = pd.DataFrame(se_toxin_df).reset_index()

# Preview DataFrame
se_toxin_df.head()
```

```python
toxin_pivot_df = toxin_df.pivot(index="YEAR", columns="PROGRAM", values="SAMPLE_VALUE_REPORTED")
se_toxin_df = se_toxin_df.pivot(index="YEAR", columns="PROGRAM", values="SAMPLE_VALUE_REPORTED")
# Preview that Reformatting worked
toxin_pivot_df
```

```python
fig, ax = plt.subplots()

ax.errorbar(toxin_pivot_df.index, toxin_pivot_df["COMMUNITY-SCALE MONITORING"], se_toxin_df["COMMUNITY-SCALE MONITORING"], fmt="o", color="r", linestyle="--", linewidth=0.5, label='COMMUNITY-SCALE MONITORING')
ax.errorbar(toxin_pivot_df.index, toxin_pivot_df["CSN SUPPLEMENTAL"], se_toxin_df["CSN SUPPLEMENTAL"], fmt="^", color="b", linestyle="--", linewidth=0.5, label='CSN SUPPLEMENTAL')
ax.errorbar(toxin_pivot_df.index, toxin_pivot_df["IMPROVE"], se_toxin_df["IMPROVE"], fmt="s", color="g", linestyle="--", linewidth=0.8, label='IMPROVE')
ax.errorbar(toxin_pivot_df.index, toxin_pivot_df["PM2.5 SPECIATION NETWORK"], se_toxin_df["PM2.5 SPECIATION NETWORK"], fmt="D", color="black", linestyle="--", linewidth=0.5, label='PM2.5 SPECIATION NETWORK')
ax.errorbar(toxin_pivot_df.index, toxin_pivot_df["SLAMS"], se_toxin_df["SLAMS"], fmt="D", color="yellow", linestyle="--", linewidth=0.5, label='SLAMS')
leg = ax.legend()
ax.set_xlabel("Years")
ax.set_ylabel("SAMPLE_VALUE_REPORTED")
ax.set_title("Programs Throughout the Years")
plt.grid()


plt.show()
```

```python
from citipy import citipy
coordinates = []
for index, row in chromium_df.iterrows(): 
    if (row['MONITOR_LATITUDE'], row['MONITOR_LONGITUDE']) not in coordinates:
        coordinates.append((row['MONITOR_LATITUDE'],row['MONITOR_LONGITUDE']))
# coordinates
cities = []
for coordinate_pair in coordinates:
    lat, lon = coordinate_pair
    cities.append(citipy.nearest_city(lat, lon))
# cities
city_names = []
for city in cities:
    city_names.append(city.city_name)
city_names
```

```python
city_data = pd.DataFrame({'Coordinates':coordinates,'City Name': city_names})
city_data
city_index = []
for index, row in chromium_df.iterrows():
    for cindex, crow in city_data.iterrows():
        if crow['Coordinates'][1] == row['MONITOR_LONGITUDE']:
            city_index.append(crow['City Name'])
            continue
chromium_df['City'] = city_index            
chromium_df
```

```python
grouped_city= chromium_df.groupby(["City"])


city = pd.DataFrame(grouped_city["SAMPLE_VALUE_REPORTED"].max())

city
```
